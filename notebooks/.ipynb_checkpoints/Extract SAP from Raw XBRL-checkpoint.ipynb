```python
xbrl_section = "SummaryOfSignificantAccountingPoliciesTextBlock"
xbrl_section_list = ['SummaryOfSignificantAccountingPoliciesTextBlock', 
                     'SignificantAccountingPoliciesTextBlock',
                    'BasisOfPresentationAndSignificantAccountingPoliciesTextBlock',
                    'BusinessDescriptionAndAccountingPoliciesTextBlock',
                    'OrganizationConsolidationAndPresentationOfFinancialStatementsDisclosureAndSignificantAccountingPoliciesTextBlock',
                     'OrganizationConsolidationBasisOfPresentationBusinessDescriptionAndAccountingPoliciesTextBlock',
                    'NewAccountingPronouncementsPolicyPolicyTextBlock']
```

```python
def extract_xml_zip(input_filename, output_dir):
    ## Extract zip file of XML submission
    print("Input file: ", input_filename)

    if not os.path.isdir(output_dir):
        os.makedirs(output_dir)
    with zipfile.ZipFile(input_filename, 'r') as zip_ref:
        zip_ref.extractall(output_dir)

    ## Get correct XML file from directory of unzipped files
    contents = os.listdir(output_dir)
    for file in contents:
        if re.findall(r"[0-9]{8}\.xml", file):
            xml_file = os.path.join(output_dir,file)
            print("XML file to be parsed is ", xml_file)
            return xml_file
    return None
```

```python
## Parse XML file
def parse_xml_file(xml_file, target_filename, sections_list):
    try:
        it = ET.iterparse(xml_file)
        for _, el in it:
            el.tag = el.tag.split('}', 1)[1]  # strip all namespaces
        root = it.root
        for xbrl_section in xbrl_section_list:
            data=root.find(xbrl_section)
            if data is not None:
                text = data.text

                ## Remove HTML tables and tags
                soup = BeautifulSoup(text)
                #for table_tag in soup.findAll('table'):
                #    table_tag.extract()
                text = soup.text

                ## Transform character entity references
                text = re.sub('&#160;', ' ', text)
                text = re.sub('&#8211;', ' ', text)
                text = text.encode("ascii","ignore").decode("utf-8")

                ## Normalize whitespace
                text = " ".join(text.split())

                ## save SAP text
                with open(target_filename, "w") as outfile:
                    outfile.write(text)
                print("Sections written to ", target_filename, " (len {})".format(len(text)))
                break
    except:
        return
```

```python
def process(f):
    xml_zip_file = os.path.join(INPUT_DIR+"/{}/{}".format(year, quarter), f)
    xml_file = extract_xml_zip(input_filename=xml_zip_file, output_dir=TEMP_DIR)
    if xml_file is not None:
        sap_target_filename = os.path.join(OUTPUT_DIR+"/{}/{}".format(year, quarter), f.replace(".zip",".txt"))
        if not os.path.isdir(OUTPUT_DIR+"/{}/{}".format(year, quarter)):
            os.makedirs(OUTPUT_DIR+"/{}/{}".format(year, quarter))
        if not os.path.isfile(sap_target_filename):
            parse_xml_file(xml_file=xml_file, target_filename=sap_target_filename, sections_list=xbrl_section_list)

    ## Remove temporary files
    shutil.rmtree(TEMP_DIR)
```

```python
for year in range(2015,2019):
    for quarter in range(1,5):
        quarter = str(quarter).zfill(2)
        input_files = os.listdir(INPUT_DIR+"/{}/{}".format(year, quarter))
        for f in input_files:
            process(f)
```
